# 沪深300数据集的算法实现 (Daily)
对我所实现的模型的缺陷的设想：
1. 未剔除剔除黑名单股票和不可交易股票，其中黑名单包括新股（上市不满12个月）、风险警示股、待退市股和净资产为负股，不可交易股票包括停牌股票和一字板股票。
2. 未考虑26年间持仓股票（可能）遭遇退市的情况。
3. 作为每日换仓的算法，应当考虑至少万分之五的佣金和手续费。
4. 日频换仓应更偏向于短期的趋势线交易，在模型中加入包含Market Value和Book-to-Market ratio的基本面因子是否有效？同时没有考虑到短期的不正常波动（应当设置止损和止盈点？或者根据股价所在的历史价格百分位设置惩罚因子？）
5. 动量积累的路径很重要！在同一时间段内看似相同（实则由不同路径积累）的动量可能在该时间段内保持上涨的天数相差巨大（例如来源于短期的高波动）。

```python
# Define a function to distinguish the codes with invalid data.
def allNaN(df: pd.DataFrame) -> list:
    lst = []
    for code in df.columns:
        if len(df[code][~np.isnan(df[code])]) == 0:
            lst.append(code)
        else:
            continue
    return lst


# Generate a list of such codes and delete them.
nanLst = list(set(allNaN(close_df) + allNaN(return_df) + allNaN(MV_df) + allNaN(BM_df)))
for code in nanLst:
    del close_df[code]
    del return_df[code]
    del MV_df[code]
    del BM_df[code]

del nanLst
```

```python
# Deal with 4 datasets.
# The rule is, the NaNs which started since the first day of the dataset would be regarded as 
# "had not been listed yet", and hence would remain NaNs. Once a positive value appeared, 
# then regarded the following NaNs as suspensions, and hence replace those NaNs by the
# last positive value before the suspension. 

# The reason why I distinguish the return dataset from other three datasets is that, the NaNs 
# cannot be regarded as the same return of its last active trading day even it is a suspension,
# because no trading means no variation.

def MissingVal(col: pd.Series) -> pd.Series:
    sinceWhen = col[~np.isnan(col)].index[0]
    col[col.index >= sinceWhen] = col[col.index >= sinceWhen].fillna(method="ffill")
    return col

def MissingVal0(col: pd.Series) -> pd.Series:
    sinceWhen = col[~np.isnan(col)].index[0]
    col[col.index >= sinceWhen] = col[col.index >= sinceWhen].fillna(0)
    return col

close_df = close_df.apply(MissingVal, axis = 0)
MV_df = MV_df.apply(MissingVal, axis = 0)
BM_df = BM_df.apply(MissingVal, axis = 0)
return_df = return_df.apply(MissingVal0, axis = 0)
```

# 4. Calculating monthly rate of returns (risk-free interest rates do not take non-trading days into consideration, such as treasury-bonds, deposit and LPR).

```python
# Determing daily risk-free rate of returns
def RF(row) -> float:
    date = row[0]
    if date <= pd.to_datetime("2002-08-06"):
        df = deposit_interest_rate
    elif date <= pd.to_datetime("2006-10-07"):
        df = central_bank_bill
    else:
        df = shibor
    if df[date <= df.index].shape[0] != 0:
        return df[date <= df.index].iloc[0, 0]
    else:
        raise Exception("Time Interval Exceeded!")


# Determine the exact year and month for every trading days.
def GetMonthlyPeriod(row) -> pd.Period:
    row = re.split(pattern = "-", string = str(row["month"]))
    return pd.Period(row[0] + "-" + row[1])
    

# Calculate the compounded interest rate for every month
def GetMonthlyReturn(col) -> float:
    if sum(np.isnan(col)) == len(col):
        return np.nan
    else:
        return np.nanprod(col)
    
    
# Create a frame which records all the information we need for risk-free rate of returns
risk_free = pd.DataFrame(
    {"date": calender.values, "month": map(lambda x: x + pd.offsets.MonthEnd(n=0), calender.values)})
risk_free["rate"] = risk_free.apply(RF, axis = 1)


# Convert to a monthly-based dataset
risk_free["month"] = risk_free.apply(GetMonthlyPeriod, axis = 1)
risk_free.set_index("month", inplace = True)
risk_free.to_csv("C:\\Users\\asus\\Desktop\\Carhart\\monthly_data\\risk_free.csv", index = True, header = True)
print("Daily risk free interest returns")
risk_free.head()
```

```python
# Summarize data of interest rates
monthly_rf = risk_free.copy()
monthly_rf["rate"] = monthly_rf["rate"] + 1
monthly_rf = monthly_rf.reset_index().groupby("month").agg({"rate": lambda x: GetMonthlyReturn(x)-1})
del risk_free

print("Monthly-compounded risk free interest returns")
monthly_rf
```

# 5. Calculating factors
According to $\mathbf{\mathit{\mbox{On Persistence in Mutual Fund Performanc}}}$ by $\mathbf{\mbox{MarK M. Carhart}}$ himself, the model can be interpreted as the following formula:

$$
\mathbb{E}[r_{p,t}] - r_{f} = \alpha + \beta_{RMRF_{p,t}}RMRF_{p,t} + \beta_{SMB_{p,t}}SMB_{p,t} + \beta_{HML_{p,t}}HML_{p,t} + \beta_{UMD_{p,t}}UMD_{p,t}
$$

Where $r_{f}$ is the risk-free interest rate, RMRF is the market risk premium = $r_{M}-r_{f}$, and $\mathbb{E}[r_{p, t}]$ is the expected return of portfolio under our assumptions.

As a reminder, for market values, book-to-market ratios and close prices I simply used monthly averages as substites for their monthly performance, while interest rates were converted into monthly compounded interest rates.

## 5.2 The accumulated rate of returns during the past 12 months (lagged 1 months)

```python
# Create and compute a DataFrame which contains all information of momentums.
momentum = monthly_return.copy().shift(1, axis = 0)

for days in range(2, 12):
    momentum = momentum + monthly_return.copy().shift(days, axis = 0)
momentum
```

```python
# Approach weighted average total market value of each stock
def GetWeightArray(month: pd.Period) -> np.array:
    row = monthly_MV[monthly_MV.index == month].iloc[0].values
    return row / np.nansum(row)


# Approach Estimated Market Return
def GetMarketReturn(month: pd.Period) -> float:
    weight = GetWeightArray(month)
    row = monthly_return[monthly_return.index == month].iloc[0].values
    return np.nansum([weight[i]*row[i] for i in range(len(row))])

# Approach SMB factors
def GetSMB(month: pd.Period) -> float:
    row = list(monthly_MV[monthly_MV.index == month].iloc[0].values)
    weight = list(GetWeightArray(month))
    returnRow = list(monthly_return[monthly_return.index == month].iloc[0].values)
    
    # filtering stocks which are non-tradable 
    for i in range(len(returnRow)-1, -1, -1):
        if np.isnan(returnRow[i]) or np.isnan(row[i]):
            weight.pop(i)
            returnRow.pop(i)
            row.pop(i)
            
    # Determine the long and short strategy of the underlying assets,
    # based on the pre-determined quantiles.
    [lower_bound, upper_bound] = np.quantile(row, [tradePercent, 1-tradePercent])
    row, weight, returnRow = np.array(row), np.array(weight), np.array(returnRow)
    return  np.dot(weight[row <= lower_bound], returnRow[row <= lower_bound]) - \
            np.dot(weight[row >= upper_bound], returnRow[row >= upper_bound])
    
    
# Approach HML factors 
def GetHML(month: pd.Period) -> float:
    row = list(monthly_BM[monthly_BM.index == month].iloc[0].values)
    weight = list(GetWeightArray(month))
    returnRow = list(monthly_return[monthly_return.index == month].iloc[0].values)
    
    # filtering stocks which are non-tradable 
    for i in range(len(returnRow)-1, -1, -1):
        if np.isnan(returnRow[i]) or np.isnan(row[i]):
            weight.pop(i)
            returnRow.pop(i)
            row.pop(i)
            
    # Determine the long and short strategy of the underlying assets,
    # based on the pre-determined quantiles.
    [lower_bound, upper_bound] = np.quantile(row, [tradePercent, 1-tradePercent])
    row, weight, returnRow = np.array(row), np.array(weight), np.array(returnRow)
    return  np.dot(weight[row >= upper_bound], returnRow[row >= upper_bound]) - \
            np.dot(weight[row <= lower_bound], returnRow[row <= lower_bound])


# Approach UMD factors
def GetUMD(month: pd.Period) -> float:
    row = list(momentum[momentum.index == month].iloc[0].values)
    weight = list(GetWeightArray(month))
    returnRow = list(monthly_return[monthly_return.index == month].iloc[0].values)
    
    # filtering stocks which are non-tradable 
    for i in range(len(returnRow)-1, -1, -1):
        if np.isnan(returnRow[i]) or np.isnan(row[i]):
            weight.pop(i)
            returnRow.pop(i)
            row.pop(i)
            
    if row == []: 
        return np.nan
    else:
        # Determine the long and short strategy of the underlying assets,
        # based on the pre-determined quantiles.
        [lower_bound, upper_bound] = np.quantile(row, [tradePercent, 1-tradePercent])
        row, weight, returnRow = np.array(row), np.array(weight), np.array(returnRow)
        return  np.dot(weight[row >= upper_bound], returnRow[row >= upper_bound]) - np.dot(weight[row <= lower_bound], returnRow[row <= lower_bound])
```

## 5.3  Output

```python
# Create a DataFrame which contains the final four factors monthly.
fourFactor = monthly_rf.copy().rename(columns = {"rate": "RF"})

fourFactor["MKT"] = np.array(map(GetMarketReturn, fourFactor.index))
fourFactor["RMRF"] = fourFactor["MKT"] - fourFactor["RF"]
fourFactor["SMB"] = np.array(map(GetSMB, fourFactor.index))
fourFactor["HML"] = np.array(map(GetHML, fourFactor.index))
fourFactor["UMD"] = np.array(map(GetUMD, fourFactor.index))

fourFactor.to_csv("C:\\Users\\asus\\Desktop\\Carhart\\monthly_data\\Factor_Result.csv", 
                  index = True, header = True)
fourFactor
```
